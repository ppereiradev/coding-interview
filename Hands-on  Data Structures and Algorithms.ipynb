### Measuring the time difference between a recursive and interative factorial function

```python
def factorial_interative(n):
    result = 1
    for i in range(1,n+1):
        result *= i
    return result
```

```python
def factorial_recursive(n):
    if n == 0:
        return 1
    else:
        return n*factorial_recursive(n-1)
```

```python
factorial_interative(10)
```

```python
factorial_recursive(10)
```

```python
import time

start = time.time()
factorial_interative(1000)
end = time.time()
print(end - start)
```

```python
import time

start = time.time()
factorial_recursive(1000)
end = time.time()
print(end - start)
```

As we can see, the interative function is faster, also the recursive one cannot compute the factorial of 10000 because it exceeds the limit of recursion

#### SIngly linked list

```python
class SinglyLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
        self.size = 0
    
    def insert_head(self, data):
        node = Node(data)
        
        if self.head is None:
            self.head = node
            self.tail = node
        else:
            current, self.head = self.head, node
            current.add_next(node)
        
        self.size += 1
            
    def append(self, data):
        node = Node(data)
        
        if self.tail is None:
            self.head = node
            self.tail = node
        else:
            current, self.tail = self.tail, node
            current.add_next(node)
        
        self.size += 1
            
    def get_size_o_n(self):
        current = self.head
        count = 0
        while current:
            count += 1
            current = current.next
        return count
    
    def get_size_o_1(self):
        return self.size
    
    def iter(self):
        current = self.head
        while current:
            val = current.data
            current = current.next
            yield val
            
    def delete(self, data):
        if self.head is None:
            print("List is empty")
        else:
            previous = self.head
            current = self.head
            while current:
                if current.data == data:
                    previous.next, current.next, current = current.next, None, None
                    self.size -= 1
                    break
                previous, current = current, current.next
    
    def search(self, data):
        for node in self.iter():
            if node == data:
                return True
        
        return False
    
    def clear(self):
        self.head = None
        self.tail = None
                    
```

#### Doubly linked list

```python
class Node:
    def __init__(self, data=None, next=None, prev=None):
        self.data = data
        self.next = next
        self.prev = prev
        
    def add_next(self, node):
        node.prev = self
        self.next = node
        
    def traverse(self):
        current = self

        while current:
            print(current.data)
            current = current.next
    
    def __str__(self):
        return str(self.data)
```

```python
class DoublyLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
        self.size = 0
    
    def insert_head(self, data):
        node = Node(data)
        
        if self.head is None:
            self.head = node
            self.tail = node
        else:
            current, self.head = self.head, node
            current.add_next(node)
        
        self.size += 1
            
    def append(self, data):
        node = Node(data)
        
        if self.tail is None:
            self.head = node
            self.tail = node
        else:
            current, self.tail = self.tail, node
            current.add_next(node)
        
        self.size += 1
            
    def get_size_o_n(self):
        current = self.head
        count = 0
        while current:
            count += 1
            current = current.next
        return count
    
    def get_size_o_1(self):
        return self.size
    
    def iter(self):
        current = self.head
        while current:
            val = current.data
            current = current.next
            yield val
            
    def delete(self, data):
        if self.head is None:
            print("List is empty")
        else:
            previous = self.head
            current = self.head
            while current:
                if current.data == data:
                    previous.next = current.next
                    current.next.prev = previous
                    current, current.next = None, None
                    self.size -= 1
                    break
                previous, current = current, current.next
    
    def search(self, data):
        for node in self.iter():
            if node == data:
                return True
        
        return False
    
    def clear(self):
        self.head = None
        self.tail = None
                    
```

#### Circular linked list

```python
class Node:
    def __init__(self, data=None, next=None, prev=None):
        self.data = data
        self.next = next
        self.prev = prev
        
    def add_next(self, node):
        node.prev = self
        self.next = node
        
    def traverse(self):
        current = self

        while current:
            print(current.data)
            current = current.next
    
    def __str__(self):
        return str(self.data)
```

```python
class CircularLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
        self.size = 0
    
    def insert_head(self, data):
        node = Node(data)
        
        if self.head is None:
            self.head = node
            self.tail = node
        else:
            current, self.head = self.head, node
            current.add_next(node)
            self.head.prev, self.tail.next = self.tail, self.head
        
        self.size += 1
            
    def append(self, data):
        node = Node(data)
        
        if self.tail is None:
            self.head = node
            self.tail = node
        else:
            current, self.tail = self.tail, node
            current.add_next(node)
            self.head.prev, self.tail.next = self.tail, self.head
        
        self.size += 1
            
    def get_size_o_n(self):
        current = self.head
        count = 0
        while current:
            count += 1
            current = current.next
        return count
    
    def get_size_o_1(self):
        return self.size
    
    def iter(self):
        current = self.head
        prev = None
        while current != prev:
            val = current.data
            current = current.next
            prev = self.head
            yield val
            
    def delete(self, data):
        if self.head is None:
            print("List is empty")
        else:
            previous = self.head
            current = self.head
            
            while current:
                if current.data == data:
                    previous.next = current.next
                    current.next.prev = previous
                    self.size -= 1
                    break
                # break loop if it runs all nodes
                if current.next == self.head:
                    break
                previous, current = current, current.next
    
    def search(self, data):
        for node in self.iter():
            if node == data:
                return True
        
        return False
    
    def clear(self):
        self.head = None
        self.tail = None
```

#### Stack

```python
class Node:
    def __init__(self, data=None):
        self.data = data
        self.prev = None
```

```python
class Stack:
    def __init__(self):
        self.top = None
        self.size = 0
        
    def push(self, data):
        node = Node(data)
        if self.top is None:
            self.top = node
        else:
            node.prev, self.top = self.top, node
        
        self.size += 1
    
    def pop(self):
        if self.top is None:
            return None
        else:
            node, self.top = self.top, self.top.prev
            self.size -= 1
            return(node.data)
    
    def peek(self):
        if self.top is None:
            return None
        else:
            return self.top.data
```

```python
def check_brackets(statement):
    
    stack = Stack()

    for ch in statement:
        if ch in ("{", "[", "("):
            stack.push(ch)
        if ch in ("}", "]", ")"):
            last = stack.pop()
            if last == "{" and ch == "}":
                continue
            elif last == "[" and ch == "]":
                continue
            elif last == "(" and ch == ")":
                continue
            else:
                return False
        
    if stack.size > 0:
        return False
    else:
        return True
```

#### Queue - using built-in list

```python
class QueueList:
    def __init__(self):
        self.items = []
        self.size = 0
        
    def enqueue(self, data):
        self.items.insert(0, data)
        self.size += 1
    
    def dequeue(self):
        return self.items.pop()
```

#### Queue - using two stacks

```python
class QueueTwoStackList:
    def __init__(self):
        self.in_stack = []
        self.out_stack = []
        self.size = 0
        
    def enqueue(self, data):
        self.in_stack.append(data)
        self.size += 1
    
    def dequeue(self):
        if len(self.out_stack) == 0:
            while len(self.in_stack) > 0:
                self.out_stack.append(self.in_stack.pop())
        return self.out_stack.pop()
            
```

#### Queue - using Nodes

```python
class Node:
    def __init__(self, data=None, next=None, prev=None):
        self.data = data
        self.next = next
        self.prev = prev
```

```python
class QueueNodes():
    def __init__(self):
        self.head = None
        self.tail = None
        self.size = 0
        
    def enqueue(self, data):
        node = Node(data)
        if self.head is None:
            self.head = node
            self.tail = node
        else:
            self.tail.next, node.prev, self.tail = node, self.tail, node
        
        self.size += 1
    
    def dequeue(self):
        if self.head is None:
            return None
        elif self.size() == 1:
            node_aux = self.head
            self.head = None
            self.tail = None
            self.size -= 1
            return node_aux.data
        else:
            node_aux = self.head
            self.head = self.head.next
            self.head.prev = None
            node_aux.next = None
            self.size -= 1
            return node_aux.data        
```

### Tree Traversal

## Depth-first traversal

In depth-first traversal, we traverse the tree, starting from the root, and go deeper into the tree as much as possible on each child, and then continue to traverse to the next sibling. We use the recursive approach for tree traversal. There are three forms of depth-first traversal, namely, in-order, pre-order, and post-order.

### In-order (left sub-tree, root, right sub-tree )

```python
def inorder(self, root_node):
    current = root_node
    if current is None:
        return
    self.inorder(current.left)
    print(current.data)
    self.inorder(current.right)
```

### Pre-order (root, left sub-tree, right sub-tree)

```python
def preorder(self, root_node):
    current = root_node
    if current is None:
        return
    print(current.data)
    self.preorder(current.left)
    self.preorder(current.right)
```

### Post-order (left sub-tree, right sub-tree, root)

```python
def postorder(self, root_node):
    current = root_node
    if current is None:
        return
    self.postorder(current.left)
    self.postorder(current.right)
    print(current.data)
```

## Breadth-first traversal

Breadth-first traversal starts from the root of the tree and then visits every node on the next level of the tree. Then, we move to the next level in the tree, and so on. This kind of tree traversal is breadth-first as it broadens the tree by traversing all the nodes in a level before going deep into the tree.

Here, we use a queue to support the breadth-first traversal.

```python
class NodeQueue:
    def __init__(self, data=None):
        self.data = data
        self.next = None

class Queue:
    def __init__(self):
        self.head = None
        self.tail = None
        self.size = 0
        
    def enqueue(self, data):
        node = NodeQueue(data)
        
        if self.head is None:
            self.head = node
            self.tail = node
            self.size += 1
        else:
            self.tail.next = node
            self.tail = node
            self.size += 1
        
        
    def dequeue(self):
        if self.head is None:
            return
        elif self.size == 1:
            current = self.head
            self.head = None
            self.tail = None
            self.size -= 1
            return current.data
        else:
            current = self.head
            self.head = self.head.next
            self.size -= 1
            return current.data
            
```

```python
class NodeTree:
    def __init__(self, data=None):
        self.data = data
        self.right = None
        self.left = None
        
class Tree:
    def __init__(self, node=None):
        self.queue = Queue()
        self.root = node
        
    def breadth_first_traversal(self):
        list_of_nodes = []
        self.queue.enqueue(self.root)
        
        while self.queue.size > 0:
            node = self.queue.dequeue()
            list_of_nodes.append(node.data)

            if node.left is not None:
                self.queue.enqueue(node.left)
            if node.right is not None:
                self.queue.enqueue(node.right)

        
        return list_of_nodes
```

## Binary search tree implementation

```python
class NodeTree:
    def __init__(self, data=None):
        self.data = data
        self.left = None
        self.right = None
        
        
class BinarySearchTree:
    def __init__(self, root=None):
        self.root = root
    
    # O(h) - h is the height of tree
    def insert(self, data):
        node = NodeTree(data)
        
        if self.root is None:
            self.root = node
        else:
            current = self.root
            parent = None
            
            while True:
                parent = current
                if node.data < parent.data:
                    current = current.left
                    if current is None:
                        parent.left = node
                        return
                else:
                    current = current.right
                    if current is None:
                        parent.right = node
                        return
    
    def get_node_with_parent(self, data):
        parent = None
        current = self.root
        
        if current is None:
            return (current, parent)
        
        while True:
            if current.data == data:
                return (current, parent)
            elif current.data > data:
                parent = current
                current = current.left
            else:
                parent = current
                current = current.right
        
        return (current, parent)
    
    # O(h) - h is the height of tree
    def delete(self, data):
        node, parent = self.get_node_with_parent(data)
        
        if node is None and parent is None:
            return False
        
        # getting to know how much children the node has
        children_count = 0
        
        if node.left and node.right:
            children_count =  2
        elif (node.left is None) and (node.right is None):
            children_count =  0
        else:
            children_count =  1
            
        # knowing how many children the node has we are able to remove the node
        if children_count == 0:
            if parent:
                if parent.right is node:
                    parent,right = None
                else:
                    parent.left = None
        
        elif children_count == 1:
            next_node = None
            if node.left:
                next_node = node.left
            else:
                next_node = node.right
            
            if parent:
                if parent.left is node:
                    parent.left = next_node
                else:
                    parent.right = next_node
            else:
                self.root = next_node
                
        else:
            parent_of_leftmost_node = node
            leftmost_node = node.right
            while leftmost_node.left:
                parent_of_leftmost_node = leftmost_node
                leftmost_node = leftmost_node.left
                
            node.data = leftmost_node.data
            
            if parent_of_leftmost_node.left is leftmost_node:
                parent_of_leftmost_node.left = leftmost_node.right
            else:
                parent_of_leftmost_node.right = leftmost_node.right
    
    # O(h) - h is the height of tree                           
    def search(self, data):
        current = self.root
        
        while True:
            if current is None:
                return None
            elif current.data == data:
                return current.data
            elif data < current.data:
                current = current.left
            elif data > current.data:
                current = current.right
    
    
    # O(h) - h is the height of tree                            
    def find_min(self):
        current = self.root
        while current.left:
            current = current.left
        
        return current.data
    
    # O(h) - h is the height of tree
    def find_max(self):
        current = self.root
        while current.right:
            current = current.right
        
        return current.data
```

## Balacing tree implementation

There are different types of self-balancing trees, such as __red-black trees, AA trees, and scapegoat trees__. These balance the tree during each operation that modifies the tree, such as insert or delete. There are also external algorithms that balance a tree. The benefits of these are that you don't need to balance the tree on every single operation and can leave balancing to the point where you need it.

```python
class NodeTree:
    def __init__(self, data=None):
        self.data = data
        self.left = None
        self.right = None
        
        
class BalancedBinarySearchTree:
    def __init__(self, root=None):
        self.root = root
    
    # O(h) - h is the height of tree
    def insert(self, data):
        node = NodeTree(data)
        
        if self.root is None:
            self.root = node
        else:
            current = self.root
            parent = None
            
            while True:
                parent = current
                if node.data < parent.data:
                    current = current.left
                    if current is None:
                        parent.left = node
                        return
                else:
                    current = current.right
                    if current is None:
                        parent.right = node
                        return
    
    def get_node_with_parent(self, data):
        parent = None
        current = self.root
        
        if current is None:
            return (current, parent)
        
        while True:
            if current.data == data:
                return (current, parent)
            elif current.data > data:
                parent = current
                current = current.left
            else:
                parent = current
                current = current.right
        
        return (current, parent)
    
    # O(h) - h is the height of tree
    def delete(self, data):
        node, parent = self.get_node_with_parent(data)
        
        if node is None and parent is None:
            return False
        
        # getting to know how much children the node has
        children_count = 0
        
        if node.left and node.right:
            children_count =  2
        elif (node.left is None) and (node.right is None):
            children_count =  0
        else:
            children_count =  1
            
        # knowing how many children the node has we are able to remove the node
        if children_count == 0:
            if parent:
                if parent.right is node:
                    parent,right = None
                else:
                    parent.left = None
        
        elif children_count == 1:
            next_node = None
            if node.left:
                next_node = node.left
            else:
                next_node = node.right
            
            if parent:
                if parent.left is node:
                    parent.left = next_node
                else:
                    parent.right = next_node
            else:
                self.root = next_node
                
        else:
            parent_of_leftmost_node = node
            leftmost_node = node.right
            while leftmost_node.left:
                parent_of_leftmost_node = leftmost_node
                leftmost_node = leftmost_node.left
                
            node.data = leftmost_node.data
            
            if parent_of_leftmost_node.left is leftmost_node:
                parent_of_leftmost_node.left = leftmost_node.right
            else:
                parent_of_leftmost_node.right = leftmost_node.right
    
    # O(h) - h is the height of tree                           
    def search(self, data):
        current = self.root
        
        while True:
            if current is None:
                return None
            elif current.data == data:
                return current.data
            elif data < current.data:
                current = current.left
            elif data > current.data:
                current = current.right
    
    
    # O(h) - h is the height of tree                            
    def find_min(self):
        current = self.root
        while current.left:
            current = current.left
        
        return current.data
    
    # O(h) - h is the height of tree
    def find_max(self):
        current = self.root
        while current.right:
            current = current.right
        
        return current.data
```

```python
class HashItem:
    def __init__(self, key, value):
        self.key = key
        self.value = value
        
class HashTable:
    def __init__(self):
        self.size = 256
        self.slots = [None for i in range(self.size)]
        self.count = 0
        
    def _hash(self, key):
        mult = 1
        hv = 0
        for ch in key:
            hv += mult * ord(ch)
            mult += 1
        return hv % self.size
    
    def put(self, key, value):
        item = HashItem(key, value)
        h = self._hash(key)
        while self.slots[h] is not None:
            if self.slots[h].key is key:
                break
            h = (h + 1) % self.size
            
        if self.slots[h] is None:
            self.count += 1
            self.slots[h] = item
    
    def get(self, key):
        h = self._hash(key) # computer hash for the given key
        
        while self.slots[h] is not None:
            if self.slots[h].key is key:
                return self.slots[h].value
            h = (h + 1) % self.size
        return None
    
    # __setitem__ is useful for using [] to set values
    def __setitem__(self, key, value):
        self.put(key, value)
    # __getitem__ is useful for using [] to get values
    def __getitem__(self, key):
        return self.get(key)
```

```python
class NodeQueue:
    def __init__(self, vertex=None):
        self.vertex = vertex
        self.next = None

class QueueAuxGraph:
    def __init__(self):
        self.head = None
        self.tail = None
        self.size = 0
        
    def enqueue(self, vertex):
        node = NodeQueue(vertex)
        
        if self.head is None:
            self.head = node
            self.tail = node
            self.size += 1
        else:
            self.tail.next = node
            self.tail = node
            self.size += 1
    
    def dequeue(self):
        if self.head is None:
            return None
        elif self.size == 1:
            current = self.head
            self.head = None
            self.tail = None
            self.size -= 1
            return current.vertex
        else:
            current = self.head
            self.head = self.head.next
            self.size -= 1
            return current.vertex
```

```python
def breadth_first_search(graph, root):
    visited_vertices = []
    graph_queue = QueueAuxGraph()
    graph_queue.enqueue(root)
    visited_vertices.append(root)
    
    while graph_queue.size > 0:
        vertex = graph_queue.dequeue()
        adj_nodes = graph[vertex]
        
        remaining_elements = set(adj_nodes).difference(set(visited_vertices))
        if len(remaining_elements) > 0:
            for elem in sorted(remaining_elements):
                visited_vertices.append(elem)
                graph_queue.enqueue(elem)
    
    return visited_vertices
```

```python
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.l = None
        self.r = None
        self.h = 1

class AVLTree:

    def insert(self, root, key):

        if not root:
            return TreeNode(key)
        elif key < root.value:
            root.l = self.insert(root.l, key)
        else:
            root.r = self.insert(root.r, key)

        root.h = 1 + max(self.getHeight(root.l),self.getHeight(root.r))

        b = self.getBal(root)

        if b > 1 and key < root.l.value:
            return self.rRotate(root)

        if b < -1 and key > root.r.value:
            return self.lRotate(root)

        if b > 1 and key > root.l.value:
            root.l = self.lRotate(root.l)
            return self.rRotate(root)

        if b < -1 and key < root.r.value:
            root.r = self.rRotate(root.r)
            return self.lRotate(root)

        return root

    def lRotate(self, z):

        y = z.r
        T2 = y.l

        y.l = z
        z.r = T2

        z.h = 1 + max(self.getHeight(z.l),self.getHeight(z.r))
        y.h = 1 + max(self.getHeight(y.l),self.getHeight(y.r))

        return y

    def rRotate(self, z):

        y = z.l
        T3 = y.r

        y.r = z
        z.l = T3

        z.h = 1 + max(self.getHeight(z.l),self.getHeight(z.r))
        y.h = 1 + max(self.getHeight(y.l),self.getHeight(y.r))

        return y

    def getHeight(self, root):
        if not root:
            return 0
        
        return root.h

    def getBal(self, root):
        if not root:
            return 0

        return self.getHeight(root.l) - self.getHeight(root.r)

    def preOrder(self, root):

        if not root:
            return

        print("{0} ".format(root.value), end="")
        self.preOrder(root.l)
        self.preOrder(root.r)

Tree = AVLTree()
root = None

root = Tree.insert(root, 1)
root = Tree.insert(root, 2)
root = Tree.insert(root, 3)
root = Tree.insert(root, 4)
root = Tree.insert(root, 5)
root = Tree.insert(root, 6)

# Preorder Traversal
print("Preorder traversal of the","constructed AVL tree is")
Tree.preOrder(root)
print()
```

## Dijkstra's Algorithm

```python
'''
Using PriorityQueue from the queue module
'''
from queue import PriorityQueue

class Graph:
    def __init__(self, num_of_vertices):
        self.v = num_of_vertices
        self.edges = [[-1 for i in range(num_of_vertices)] for j in range(num_of_vertices)]
        self.visited = []
        
    def add_edge(self, u, v, weight):
        self.edges[u][v] = weight
        self.edges[v][u] = weight
        
    def dijkstra(self, start_vertex):
        D = {v:float('inf') for v in range(self.v)}
        D[start_vertex] = 0
    
        pq = PriorityQueue()
        pq.put((0, start_vertex))
    
        while not pq.empty():
            (dist, current_vertex) = pq.get()
            self.visited.append(current_vertex)
    
            for neighbor in range(self.v):
                if self.edges[current_vertex][neighbor] != -1:
                    distance = self.edges[current_vertex][neighbor]
                    if neighbor not in self.visited:
                        old_cost = D[neighbor]
                        new_cost = D[current_vertex] + distance
                        if new_cost < old_cost:
                            pq.put((new_cost, neighbor))
                            D[neighbor] = new_cost
        return D
```

## Priority Queues  and Heaps

Appending operation:
add -> 4
heap -> [0,5,6,7,8,9,10]
heap -> [0,5,6,7,8,9,10,4] 4//2 = 2

```python
class MinHeap:
    def __init__(self):
        self.heap = [0]
        self.size = 0
        
    def insert(self, value):
        self.heap.append(value)
        self.size += 1
        self.arrange(self.size)
        
    def arrange(self, idx):
        while idx//2 > 0:
            if self.heap[idx] < self.heap[idx//2]:
                self.heap[idx], self.heap[idx//2] = self.heap[idx//2], self.heap[idx]
            idx //= 2
            
    def _minindex(self, idx):
        if idx*2 + 1 > self.size:
            return idx*2
        elif self.heap[idx*2] < self.heap[idx*2+1]:
            return idx*2
        else:
            return idx*2 + 1
    
    def sink(self, idx):
        while idx*2 <= self.size:
            mi = self._minindex(idx)
            
            if self.heap[idx] > self.heap[mi]:
                self.heap[idx], self.heap[mi] = self.heap[mi], self.heap[idx]
            idx = mi
    
    def pop(self):
        value = self.heap[1]
        self.heap[1] = self.heap[self.size]
        self.size -= 1
        self.heap.pop()
        self.sink(1)
        return value
```

# Chapter 9 - Searching

```python
def binary_search(sequence, item):    
    start = 0
    end = len(sequence) - 1
    
    while start < end:
        midpoint_idx = start + (end - start) // 2
        if sequence[midpoint_idx] == item:
            return midpoint_idx
        elif item < sequence[midpoint_idx]:
            end = midpoint_idx - 1
        else:
            start = midpoint_idx + 1
            
    return None
```

```python
def merge_sort(arr):
    if len(arr) > 1:
        left = arr[:len(arr)//2]
        right = arr[len(arr)//2:]
        
        merge_sort(left)
        merge_sort(right)
        
        # index of left, right and arr consecutively
        i = j = k = 0
        
        while i < len(left) and j < len(right):
            if left[i] < right[j]:
                arr[k] = left[i]
                i += 1
                
            else:
                arr[k] = right[j]
                j += 1
                
            k += 1
            
        while i < len(left):
            arr[k] = left[i]
            i += 1
            k += 1
        
        while j < len(right):
            arr[k] = right[j]
            j += 1
            k += 1
```

## Quick sort

```python
def quick_sort_recursive(sequence):
    if len(sequence) <= 1:
        return sequence
    else:
        pivot = sequence.pop()
        
    items_lower = []
    items_greater = []
    
    for item in sequence:
        if item < pivot:
            items_lower.append(item)
        else:
            items_greater.append(item)
    return quick_sort_recursive(items_lower) + [pivot] + quick_sort_recursive(items_greater)
```

```python
class HeapMin:
    def __init__(self):
        self.heap = [0]
        self.size = 0
        
    def insert(self, item):
        self.heap.append(item)
        self.size += 1
        self._arrange(self.size)
        
    def _arrange(self, idx):
        while idx//2 > 0:
            if self.heap[idx//2] > self.heap[idx]:
                self.heap[idx], self.heap[idx//2] = self.heap[idx//2], self.heap[idx]
            idx //= 2
    
    def pop(self):
        value = self.heap[1]
        self.heap[1] = self.heap[self.size]
        self.size -= 1
        self.heap.pop()
        self._sink(1)
        return value
    
    def _sink(self, idx):
        while idx*2 <= self.size:
            min_idx = self._minindex(idx)
            
            if self.heap[min_idx] < self.heap[idx]:
                self.heap[idx], self.heap[min_idx] = self.heap[min_idx], self.heap[idx]
            idx = min_idx
    
    def _minindex(self, idx):
        if idx*2 + 1 > self.size:
            return idx*2
        elif self.heap[idx*2] < self.heap[idx*2+1]:
            return idx*2
        else:
            return idx*2 + 1        
        
    def sort_list(self):
        sorted_list = []
        for elem in range(self.size):
            sorted_list.append(self.pop())
            
        return sorted_list
```

## Quick Selection

```python
def _helper_quick_selection(sequence, kth_smallest):
    if len(sequence) <= 1:
        return sequence
    else:
        pivot = sequence.pop()
    
    items_lower = []
    items_greater = []
    
    for item in sequence:
        if item < pivot:
            items_lower.append(item)
        else:
            items_greater.append(item)
    
    if kth_smallest < len(items_lower):
        return _helper_quick_selection(items_lower, kth_smallest) + [pivot] + items_greater
    elif kth_smallest > len(items_lower):
        return items_lower + [pivot] + _helper_quick_selection(items_greater, kth_smallest)
    else:
        return items_lower + [pivot] + items_greater
    
def quick_selection(sequence, kth_smallest):
    half_sorted = _helper_quick_selection(sequence, kth_smallest)
    return half_sorted[kth_smallest - 1]
    
```

### Rabin-Karp algorithm

It uses a hash function to calculate the substrings hash values then compare, so it does not need to compare each letter of the substrings. If the hash matches, the algorithm needs to verify if the substring matches each letter, because as we know the hash sometimes generates the same value for two different inputs.

```python
# preprocessing
def generate_hash(text, pattern):
    # stores unicode value of each character in text
    ord_text = [ord(i) for i in text]
    
    # stores unicode value of each character in pattern
    ord_pattern = [ord(j) for j in pattern]
    
    # stores length of the text
    len_text = len(text)
    
    # stores length of the pattern
    len_pattern = len(pattern)
    
    hash_pattern = sum(ord_pattern)
    
    #stores the length of new array that will contain the hash values of text
    len_hash_array = len_text - len_pattern + 1
    
    # Initialize all the values in the array to 0.
    hash_text = [0]*(len_hash_array)
    
    for i in range(0, len_hash_array):
        if i == 0:
            # initial value of hash function
            hash_text[i] = sum(ord_text[:len_pattern])
            
        else:
            # calculating next hash value using previous value
            hash_text[i] = ((hash_text[i-1] - ord_text[i-1]) + ord_text[i+len_pattern-1])
    
    # return the hash values
    return hash_text, hash_pattern
```

```python
def rabin_karp_matcher(text, pattern):
    
    # convert text and pattern into string format
    text = str(text)
    pattern = str(pattern)
    
    # generate hash values using generate_hash function
    hash_text, hash_pattern = generate_hash(text, pattern)
    
    print(hash_text, hash_pattern)
    
    # length of text
    len_text = len(text)
    
    # length of pattern
    len_pattern = len(pattern)

    # checks if pattern is present atleast once or not at all
    flag = False 
    
    for i in range(len(hash_text)):
        if hash_text[i] == hash_pattern:
            count = 0
            
            # comparing patten and substring character by character
            for j in range(len_pattern):
                if pattern[j] == text[i+j]:
                    count += 1
                else:
                    break
            
            # Pattern is found in the text
            if count == len_pattern:        
                flag = True
                print("Pattern occurs at index", i)
            
    # Pattern doesn't match even once.
    if not flag:
        print("Pattern is not at all present in the text")
    
```

## Divide and Conquer

### Merge-sort

```python
def merge_sort(sequence):
    if len(sequence) == 1:
        return sequence
    
    mid = len(sequence) // 2
    left = sequence[:mid]
    right = sequence[mid:]

    # Recursive call on each half    
    half_a = merge_sort(left)
    half_b = merge_sort(right)
    
    result = merge(half_a, half_b)
    
    return result

def merge(first_sublist, second_sublist):
    # Two iterators for traversing the two halves
    i = j = 0
    merged_list = []
    
    while i < len(first_sublist) and j < len(second_sublist):
        if first_sublist[i] < second_sublist[j]:
            merged_list.append(first_sublist[i])
            i += 1
        else:
            merged_list.append(second_sublist[j])
            j += 1
            
    while i < len(first_sublist):
        merged_list.append(first_sublist[i])
        i += 1
        
    while j < len(second_sublist):
        merged_list.append(second_sublist[j])
        j += 1
        
    return merged_list
```

```python
def coin_counting(total, list_coins):
    sorted_denominations = sorted(list_coins, reverse=True)
    
    number_of_denoms = []
    
    for i in sorted_denominations:
        div = total // i
        total = total % i
        if div > 0:
            number_of_denoms.append((i, div))
    
    return number_of_denoms
```

```python
def optimal_small_change(total_amount, denom):
    sorted_denominations = sorted(denom, reverse=True)
    series = []

    for j in range(len(sorted_denominations)):
        term = sorted_denominations[j:]
        number_of_denoms = []
        local_total = total_amount
        
        for i in term:
            div = local_total // i
            local_total = local_total % i
            if div > 0:
                number_of_denoms.append((i, div))
        
        series.append(number_of_denoms)
    
    min_amt_coin = float('inf')
    best_serie = (0,)
    for serie in series:
        amt_coin = 0
        for comb in serie:
            amt_coin += comb[1]
        
        if amt_coin < min_amt_coin:
            min_amt_coin = amt_coin
            best_serie = serie
    
    return "Minimum Amout of Coins: {} - Combination: {}".format(min_amt_coin, best_serie)
```

```python
print(optimal_small_change(12,[1,5,8]))
```

### Shortest path algorithm - Dijkstra's Algorithm

```python
graph = dict()
graph['A'] = {'B': 5, 'D': 9, 'E': 2}
graph['B'] = {'A': 5, 'C': 2}
graph['C'] = {'B': 2, 'D': 3}
graph['D'] = {'A': 9, 'F': 2, 'C': 3}
graph['E'] = {'A': 2, 'F': 3}
graph['F'] = {'E': 3, 'D': 2}
```

```python
table = dict()
table = {
'A': [0, None],
'B': [float("inf"), None],
'C': [float("inf"), None],
'D': [float("inf"), None],
'E': [float("inf"), None],
'F': [float("inf"), None],
}
```

```python
# to avoid magic numbers, therefore we defined
# constants
DISTANCE = 0
PREVIOUS_NODE = 1
INFINITY = float('inf')

def find_shortest_path(graph, table, origin):
    visited_nodes = []
    current_node = origin
    starting_node = origin
    
    while True:
        adjacent_nodes = graph[current_node]
        if set(adjacent_nodes).issubset(set(visited_nodes)):
        # Nothing here to do. All adjacent nodes have been visited.
            pass
        else:
            unvisited_nodes = set(adjacent_nodes).difference(set(visited_nodes))
            for vertex in unvisited_nodes:
                distance_from_starting_node = get_shortest_distance(table, vertex)
                
                if distance_from_starting_node == INFINITY and current_node == starting_node:
                    total_distance = get_distance(graph, vertex, current_node)
                else:
                    total_distance = get_shortest_distance (table, current_node) + get_distance(graph, current_node, vertex)
                
                if total_distance < distance_from_starting_node:
                    set_shortest_distance(table, vertex,total_distance)
                    set_previous_node(table, vertex, current_node)

        visited_nodes.append(current_node)
            
        if len(visited_nodes) == len(table.keys()):
            break
            
        current_node = get_next_node(table,visited_nodes)
    
    return table
    
# helpers
def get_shortest_distance(table, vertex):
    shortest_distance = table[vertex][DISTANCE]
    return shortest_distance

def set_shortest_distance(table, vertex, new_distance):
    table[vertex][DISTANCE] = new_distance
    
def set_previous_node(table, vertex, previous_node):
    table[vertex][PREVIOUS_NODE] = previous_node

def get_distance(graph, first_vertex, second_vertex):
    return graph[first_vertex][second_vertex]

def get_next_node(table, visited_nodes):
    unvisited_nodes = list(set(table.keys()).difference(set(visited_nodes)))
    assumed_min = table[unvisited_nodes[0]][DISTANCE]
    min_vertex = unvisited_nodes[0]
    
    for node in unvisited_nodes:
        if table[node][DISTANCE] < assumed_min:
            assumed_min = table[node][DISTANCE]
            min_vertex = node
    
    return min_vertex
```

```python
shortest_distance_table = find_shortest_path(graph, table, 'A')
for k in sorted(shortest_distance_table):
    print("{} - {}".format(k,shortest_distance_table[k]))
```

# 1 - Two Sum (Easy)

Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target.

You may assume that each input would have exactly one solution, and you may not use the same element twice.

You can return the answer in any order.

__Example 1:__

__Input:__

nums = [2,7,11,15], target = 9

__Output:__ [0,1]

__Explanation:__ Because nums[0] + nums[1] == 9, we return [0, 1].

__Example 2:__

__Input:__

nums = [3,2,4], target = 6

__Output:__ [1,2]

__Example 3:__

__Input:__ 

nums = [3,3], target = 6

__Output:__ [0,1]
 

__Follow-up:__ Can you come up with an algorithm that is less than O(n2) time complexity?

```python
def find_two_sum(arr, target):
    temp = {arr[i]:i for i in range(len(arr))}
    
    for i in range(len(arr)):
        if target-arr[i] in temp:
            return(i, temp[target-arr[i]])
```

```python
nums = [2,7,11,15]
target = 9

find_two_sum(nums, target)
```

# 2 - Insert Interval (Medium)

You are given an array of non-overlapping intervals intervals where intervals[i] = [starti, endi] represent the start and the end of the ith interval and intervals is sorted in ascending order by starti. You are also given an interval newInterval = [start, end] that represents the start and end of another interval.

Insert newInterval into intervals such that intervals is still sorted in ascending order by starti and intervals still does not have any overlapping intervals (merge overlapping intervals if necessary).

Return intervals after the insertion.

 

__Example 1:__

__Input:__

intervals = [[1,3],[6,9]], newInterval = [2,5]

__Output:__ [[1,5],[6,9]]

__Example 2:__

__Input:__ 

intervals = [[1,2],[3,5],[6,7],[8,10],[12,16]], newInterval = [4,8]

__Output:__ [[1,2],[3,10],[12,16]]

__Explanation:__ Because the new interval [4,8] overlaps with [3,5],[6,7],[8,10].

```python
# TC: O(n), SC: O(n)
def insert_interval(intervals, new_interval):
    result = []
    
    for i in range(len(intervals)):
        if new_interval[1] < intervals[i][0]:
            result.append(new_interval)
            return result + intervals[i:]
        
        elif new_interval[0] > intervals[i][1]:
            result.append(intervals[i])
        
        else:
            new_interval = [min(new_interval[0], intervals[i][0]),
                           max(new_interval[1], intervals[i][1])]
        
    result.append(new_interval)
    return result
```

```python
intervals = [[1,2],[3,5],[6,7],[8,10],[12,16]]
new_interval = [4,8]

insert_interval(intervals, new_interval)
```

# 3 - Number swapper

Write a function to swap a number in place (that isf without temporary variables).


```python
def swap(num1, num2):
    num1 = num1 + num2
    num2 = num1 - num2
    num1 = num1 - num2
    
    return num1, num2
```

For the last line of text, it should be left-justified and no extra space is inserted between words.

__Note:__

```python
"""
words = ["This", "is", "an", "example", "of", "text", "justification."]
len   = [   4,    2,    2,       7,       2,     4,          14       ]
                                                              i
                                                                            j
spaces = 3
len_line = 32
"""
def middle_justify(line, max_width):
    space_spot = len(line) - 1
    
    spaces_needed = (max_width - sum(len(word) for word in line)) // space_spot
    extra_spaces = (max_width - sum(len(word) for word in line)) % space_spot
    
    for spot in range(space_spot):
        line[spot] += " " * spaces_needed + (" " if extra_spaces > 0 else "")
        extra_spaces -= 1
    return "".join(line)

def left_justify(line, max_width):
    space_spot = len(line) - 1
    spaces_needed = max_width - sum(len(word) for word in line) - space_spot
    
    return " ".join(line) + " " * spaces_needed
    
def text_justify(words, max_width):
    text = []
    i = 0
    j = i + 1
    line_length = 0
    while j < len(words):
        if line_length == 0:
            line_length = len(words[i]) + len(words[j])
        else:
            num_spaces = j - i - 1
            line_length += len(words[j]) + num_spaces
        
        if line_length > max_width:
            text.append(words[i:j])
            line_length = 0
            i = j
            j = i + 1
            continue
        
        j += 1
        
    text.append(words[i:j])
    
    for i in range(len(text)):
        if len(text[i]) == 1 or i == len(text) - 1:
            text[i] = left_justify(text[i], max_width)
        elif len(text[i]) > 1:
            text[i] = middle_justify(text[i], max_width)
    
    
    return text
```

# 5 - Minimum Window Substring (Hard)

Given two strings s and t of lengths m and n respectively, return the minimum window substring of s such that every character in t (including duplicates) is included in the window. If there is no such substring, return the empty string "".

The testcases will be generated such that the answer is unique.

A substring is a contiguous sequence of characters within the string.
 
__Example 1:__

Input: s = "ADOBECODEBANC", t = "ABC"

Output: "BANC"

Explanation: The minimum window substring "BANC" includes 'A', 'B', and 'C' from string t.


__Example 2:__

Input: s = "a", t = "a"

Output: "a"

Explanation: The entire string s is the minimum window.


__Example 3:__

Input: s = "a", t = "aa"

Output: ""

Explanation: Both 'a's from t must be included in the window. Since the largest window of s only has one 'a', return empty string.
  

__Follow up:__ Could you find an algorithm that runs in O(m + n) time?

```python
"""

  i
          j
A D O B E C O D E B A N C

ABC

min_window = 4

"""
def is_valid(s_hash, t_hash):
    for key in t_hash:
        if s_hash[key] < t_hash[key]:
            return False
    return True

def find_min_window(s, t):
    start = end = 0
    min_window = float("inf")    
    min_substring = ""
    s_hash = {i:0 for i in t}
    t_hash = {}
    for i in t:
        if i in t_hash:
            t_hash[i] += 1
        else:
            t_hash[i] = 1
            
    while start < len(s) and end <= len(s):
        if is_valid(s_hash, t_hash):
            if s[start] in s_hash:
                s_hash[s[start]] -= 1
            if min_window > end-start:
                min_window = end - start
                min_substring = s[start:end]
            
            start += 1
            continue
            
        elif end < len(s) and s[end] in s_hash:
            s_hash[s[end]] += 1
        
        end += 1
        
    return min_window, min_substring
```

```python
s = "ADOBECODEBANC"
t = "ABC"
find_min_window(s, t)
```

For this problem, when there is a tie (i.e., two or more keys with the same frequency), the least recently used key would be invalidated.
To determine the least frequently used key, a use counter is maintained for each key in the cache. The key with the smallest use counter is the least frequently used key.

When a key is first inserted into the cache, its use counter is set to 1 (due to the put operation). The use counter for a key in the cache is incremented either a get or put operation is called on it.

__Example 1:__

Input
["LFUCache", "put", "put", "get", "put", "get", "get", "put", "get", "get", "get"]

Output
[null, null, null, 1, null, -1, 3, null, -1, 3, 4]

__Explanation__

// cnt(x) = the use counter for key x

// cache=[] will show the last used order for tiebreakers (leftmost element is  most recent)

LFUCache lfu = new LFUCache(2);

lfu.put(1, 1);   // cache=[1,_], cnt(1)=1

lfu.put(2, 2);   // cache=[2,1], cnt(2)=1, cnt(1)=1

                 // cache=[1,2], cnt(2)=1, cnt(1)=2

lfu.put(3, 3);   // 2 is the LFU key because cnt(2)=1 is the smallest, invalidate 2.
                 // cache=[3,1], cnt(3)=1, cnt(1)=2

lfu.get(2);      // return -1 (not found)

lfu.get(3);      // return 3
                 // cache=[3,1], cnt(3)=2, cnt(1)=2

lfu.put(4, 4);   // Both 1 and 3 have the same cnt, but 1 is LRU, invalidate 1.
                 // cache=[4,3], cnt(4)=1, cnt(3)=2

lfu.get(1);      // return -1 (not found)

lfu.get(3);      // return 3
                 // cache=[3,4], cnt(4)=1, cnt(3)=3

lfu.get(4);      // return 4
                 // cache=[4,3], cnt(4)=2, cnt(3)=3

```python
class BinaryTree:
    class Node:
        def __init__(self, data=None):
            self.data = data
            self.left = None
            self.right = None
    
    def __init__(self, root=None):
        self.root = root
    
    def insert_interative(self, data):
        node = self.Node(data)
        
        if self.root is None:
            self.root = node
        else:
            parent = self.root
            
            while True:
                if node.data <= parent.data:
                    if parent.left is None:
                        parent.left = node
                        break
                    else:
                        parent = parent.left
                else:
                    if parent.right is None:
                        parent.right = node
                        break
                    else:
                        parent = parent.right
    
    def insert_recursion(self, data):
        def _insert(node, data):
            if node is None:
                return self.Node(data)
            
            if data <= node.data:
                node.left = _insert(node.left, data)
            elif data > node.data:
                node.right = _insert(node.right, data)
            
            return node
        
        if self.root is None:
            self.root = self.Node(data)
        else:
            self.root = _insert(self.root, data)
            
        
    def inorder_traversal(self):        
        def _inorder_traversal(node, result):
            if node is None:
                return
            
            _inorder_traversal(node.left, result)
            result.append(node.data)
            _inorder_traversal(node.right, result)
            
            
        result = []
        _inorder_traversal(self.root, result)
        
        return result
    
    def preorder_traversal(self):
        def _preorder_traversal(node, result):
            if node is None:
                return
            
            result.append(node.data)
            _preorder_traversal(node.left, result)
            _preorder_traversal(node.right, result)
            
        result = []
        _preorder_traversal(self.root, result)
        
        return result
            
    
    def postorder_traversal(self):
        def _postorder_traversal(node, result):
            if node is None:
                return
            
            _postorder_traversal(node.left, result)
            _postorder_traversal(node.right, result)
            result.append(node.data)
            
        result = []
        _postorder_traversal(self.root, result)
        
        return result
```

# 6 - Print Root to Node Path in a Binary Tree

__Problem Statement:__ Print Root to Node Path In A Binary Tree. Write a program to print path from root to a given node in a binary tree.

__Problem Description:__

We are given a binary tree T and a node V. We need to print a path from the root of the tree to the node.

__Note:__

 - No two nodes in the tree have the same data value.
 - It is assured that the node V is present and a path always exists.

```python
from collections import deque
class BinaryTree:
    class Node:
        def __init__(self, data=None):
            self.data = data
            self.left = None
            self.right = None
            
    def __init__(self, root=None):
        self.root = root
    
    def insert(self, data):
        def _insert(node, data):
            if node is None:
                return self.Node(data)
            
            if data <= node.data:
                node.left = _insert(node.left, data)
            else:
                node.right = _insert(node.right, data)
                
            return node
        
        if self.root is None:
            self.root = self.Node(data)
        else:
            self.root = _insert(self.root, data)
    
    def find_path(self, value):
        def _inorder(node, value, temp):
            if node is None:
                return False
            elif node.data == value:
                temp.append(node.data)
                return True
            
            if _inorder(node.left, value, temp):
                temp.append(node.data)
            if _inorder(node.right, value, temp):
                temp.append(node.data)
            
        temp = []
        _inorder(self.root, value, temp)
        if len(temp) > 0:
            temp.append(self.root.data)
        
        result = []
        for i in range(len(temp)):
            result.append(temp.pop())
            
        return result
    
    def breadth_first(self):
        queue = deque()
        queue.append(self.root)
        
        result = []
        
        while len(queue) > 0:
            curr = queue.popleft()
            
            if curr.left is not None:
                queue.append(curr.left)
            if curr.right is not None:
                queue.append(curr.right)
        
            result.append(curr.data)
            
        return result
    
    # zig zag traversal
    def zig_zag_traversal(self):
        queue = deque()
        queue.append(self.root)
        left_to_right = True
        result = []
        
        while len(queue) > 0:
            row = deque()
            for i in range(len(queue)):
                cur = queue.popleft()
                
                if left_to_right:
                    row.append(cur.data)
                else:
                    row.appendleft(cur.data)
                    
                if cur.left is not None:
                    queue.append(cur.left)
                if cur.right is not None:
                    queue.append(cur.right)
                
            result += row
            left_to_right = not left_to_right
            
        return result
        
    
    def reverse(self):
        def _reverse(node):
            if node is None:
                return
            
            temp = node.left
            node.left = node.right
            node.right = temp
            
            _reverse(node.left)
            _reverse(node.right)
        
        _reverse(self.root)
```

# 7 - Lowest Common Ancestor for two given Nodes

__Problem Statement:__ Given a binary tree, Find the Lowest Common Ancestor for two given Nodes (x,y).

Lowest Common Ancestor(LCA): The lowest common ancestor is defined between two nodes x and y as the lowest node in T that has both x and y as descendants (where we allow a node to be a descendant of itself.

```python
def lca_helper(root, p, q):
    if root is None or root.data == p or root.data == q:
        return root
    
    left = lca_helper(root.left, p, q)
    right = lca_helper(root.right, p, q)
    
    if left is None:
        return right
    elif right is None:
        return left
    else:
        return root
    
def lowest_common_ancestor(tree, p, q):
    node = lca_helper(tree.root, p, q)
    
    return node.data
    
```

# 11 - Check Balanced

Implement a function to check if a binary tree is balanced. For the purposes of this question, a balanced tree is defined to be a tree such that the heights of the two subtrees of any node never differ by more than one.

```python
def height(node):
    if node is None:
        return 0

    return max(height(node.left), height(node.right)) + 1

def is_balanced(node):
    if node is None:
        return True
    
    left_height = height(node.left)
    right_height = height(node.right)

    left = is_balanced(node.left)
    right = is_balanced(node.right)
    
    if abs(left_height - right_height) <= 1:
        return left and right
    
    return False
```

```python
# first common ancestor
def find_node(node, value):
    if node is None:
        return False
    
    if node.data == value:
        return True
    
    if find_node(node.left, value):
        return True
    
    if find_node(node.right, value):
        return True
    
    return False
    
def find_lca_util(node, value1, value2):
    if node is None:
        return

    if node.data == value1 or node.data == value2:
        return node.data
 
    left_lca = find_lca_util(node.left, value1, value2)
    right_lca = find_lca_util(node.right, value1, value2)
 
    if left_lca and right_lca:
        return node.data
 
    return left_lca if left_lca is not None else right_lca
```

# 16 - Check Subtree

T1 and T2 are two very large binary trees, with T1 much bigger than T2. Create an algorithm to determine if T2 is a subtree of T1. A tree T2 is a subtree of T1 if there exists a node n in T1 such that the subtree of n is identical to T2. That is, if you cut off the tree at node n, the two trees would be identical.

```python
def is_identical(node_1, node_2):
    if node_1 is None and node_2 is None:
        return True
    elif node_1 is None or node_2 is None:
        return False
    
    return (node_1.data == node_2.data and 
            is_identical(node_1.left, node_2.left) and
            is_identical(node_1.right, node_2.right))
                         
def is_subtree(node_tree, node_subtree):
    if node_subtree is None:
        return True
    elif node_tree is None:
        return False
    
    if is_identical(node_tree, node_subtree):
        return True
    
    return is_subtree(node_tree.left, node_subtree) or is_subtree(node_tree.right, node_subtree)
```

You are given a binary tree in which each node contains an integer value (which might be positive or negative). Design an algorithm to count the number of paths that sum to a given value.The path does not need to start or end at the root or a leaf, but it must go downwards (traveling only from parent nodes to child nodes).

```python
"""
                 5
               /   \
              /     \
             /       \
            3         2
          /   \     /   \   
         1     7   4     8
       /   \
      6     9
"""
def print_vector(vector, i):
    for j in range(i, len(vector)):
        print(vector[j], end=" ")
    print()

def find_path_with_sums(binary_tree, target):
    def helper(node, target, path):
        if node is None:
            return
        
        path.append(node.data)
        helper(node.left, target, path)
        helper(node.right, target, path)
        
        sum_items = 0
        for j in range(len(path) - 1, -1, -1):
            sum_items += path[j]
 
            if (sum_items == target):
                print_vector(path, j)
        
        path.pop()
        
        
    path = []
    helper(binary_tree.root, target, path)
```

# 19 - Binary Tree Longest Consecutive Sequence

```python
def find_consecutive_max(node, counter, target, max_seq):
    if node is None:
        return 0
    elif target == node.data:
        counter += 1
    else:
        counter = 1
    
    left = find_consecutive_max(node.left, counter, node.data+1, max_seq)
    right = find_consecutive_max(node.right, counter, node.data+1, max_seq)
    
    max_seq = max(counter, left, right, max_seq)
    
    return max_seq

def find_longest(binary_tree):
    max_seq = 0
    root    = binary_tree.root
    
    max_seq = find_consecutive_max(root, 0, root.data, max_seq)
    return max_seq
```
